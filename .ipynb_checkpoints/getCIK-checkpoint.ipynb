```python
import pandas as pd
from sec_edgar_downloader import Downloader
import os.path
from os import path
```

```python
cik = pd.read_csv("FINAL_COMPANY_LIST_SEC_INFO.csv")[['cik']]
cik = cik.values.tolist()
print(len(cik))
```

```python
file_location = "/Volumes/ExcellentJoe/10K/"
dl = Downloader(file_location)

num = 0
cikAsList = []
for lst in cik:
    for c in lst:
        cikAsList.append(c)
```

```python
print(len(cikAsList))
for i in range(len(cikAsList)):
    c = str(cikAsList[i])
    location = file_location + 'sec_edgar_filings' + '/' + str(c).lstrip('0') + '/'
    if (path.isdir(location)):
        print(str(c) + " was already downloaded")
    else:
        print('Started: ' + str(c))
        dl.get("10-K", str(c), 30)
        print('Downloaded: ' + c + ', ' + str(i) +' in total')
```

```python
missing = []
for c in cikAsList:
    c = str(c)
    location = file_location + 'sec_edgar_filings' + '/' + c.lstrip('0') + '/'
    if not path.exists(location):
        missing.append(c)
print('length: ' + str(len(missing)))
print(missing)
```
