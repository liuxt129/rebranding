```python
import urllib.request
import os.path
from os import path
```

```python
import requests
from bs4 import BeautifulSoup

url = "http://trademarks.reedtech.com/tmappxml.php#1884-2017"
response = requests.get(url)
# parse html
page = str(BeautifulSoup(response.content))


def getURL(page):
    """

    :param page: html of web page (here: Python home page) 
    :return: urls in that page 
    """
    start_link = page.find("a href")
    if start_link == -1:
        return None, 0
    start_quote = page.find('"', start_link)
    end_quote = page.find('"', start_quote + 1)
    url = page[start_quote + 1: end_quote]
    return url, end_quote

urls = []
while True:
    url, n = getURL(page)
    page = page[n:]
    if url:
        urls.append(url)
    else:
        break

downloadUrl = []
for url in urls:
    if url.startswith('downloads'):
        downloadUrl.append('http://trademarks.reedtech.com/' + url)
print(downloadUrl)
```

```python
location = '/pylon5/tr5pi7p/suli2020/uspto/'
num = 0
for url in downloadUrl:
    ind = url.index('apc')
    fileName = url[ind:]
    fileLocation = location + fileName
    if (path.exist(fileLocation)):
        print(fileName + " was already downloaded")
        continue
    urllib.request.urlretrieve(url, fileLocation)
    num += 1
    print('downloaded: ' + fileName + ", " + num + " files in total")
```
