```python
import pandas as pd
```

```python
case_file_cols = ['serial_no', 'abandon_dt', 'amend_reg_dt','reg_cancel_cd', 'reg_cancel_dt','file_location', 'filing_dt','registration_dt', 'renewal_dt' ]
case_file = pd.read_csv('case_file.csv', usecols=case_file_cols)
print('done reading case_file.csv')
```

```python
company_names = pd.read_csv('FINAL_COMPANY_LIST_SEC_INFO.csv', usecols=['companyName'])
print('done reading company names')
```

```python
company_names.head()
```

```python
owner_cols = ['own_addr_1', 'own_addr_city', 'own_entity_cd', 'own_name', 'own_type_cd', 'own_addr_postal', 'own_id', 'serial_no']
owner = pd.read_csv('owner.csv', usecols=owner_cols)
```

```python
# merge the owner and case_files
owner_and_cases = owner.merge(case_file, on=['serial_no', 'serial_no'])
```

```python
owner.sample(10)
```

```python
# clean a little
company_names['clean_names'] = company_names['companyName'].str.lower()
company_names['clean_names'] = company_names['clean_names'].str.strip()
owner_and_cases['clean_names'] = owner_and_cases['own_name'].str.lower()
owner_and_cases['clean_names'] = owner_and_cases['clean_names'].str.strip()

remove_the_end = r'the$'
remove_the_beg = r'^the'
owner_and_cases['clean_names'] = owner_and_cases['clean_names'].str.replace(remove_the_end, '')
owner_and_cases['clean_names'] = owner_and_cases['clean_names'].str.replace(remove_the_beg, '')
```

```python
company_names.sample(15)
```

```python
# merge our company names with theirs
merged_company_names = company_names.merge(owner_and_cases, left_on='clean_names', right_on='clean_names')
```

```python
len(merged_company_names.index)
```

```python
merged_company_names.sample(15)
```

```python
merged_company_names.to_csv('filtered_case_files_all.csv')
```

At this point, all the information from above (case file information filtered by the owner name) was written to a file as a checkpoint.

# Merging prior marks with case files

```python
prior_mark_cols = ['serial_no', 'prior_no']
prior_mark = pd.read_csv('prior_mark.csv', usecols=prior_mark_cols)
prior_mark = prior_mark.rename(columns = {'prior_no': 'prior_marks'})
```

```python
merged_all = merged_company_names.merge(prior_mark, on=['serial_no', 'serial_no'], suffixes=('_cf', '_pm'))
```
